```python
url = "https://api.census.gov/data/timeseries/healthins/sahie?get=NIC_PT,NUI_PT,NAME&for=state:*&time=2017&key={0}".format(census_api_key)
response = requests.request("GET", url)
json_to_dataframe(response)
```

```python
df_pop_by_state = json_to_dataframe(response)

# convert population to integer
df_pop_by_state['NIC_PT'] = df_pop_by_state['NIC_PT'].astype(int)

# top 5 states with largest population
df_pop_by_state_sort = df_pop_by_state\
    .sort_values(by=['NIC_PT'], ascending=False)\
    .head(5) 
print('Top 5 states with largest health population estimates:')
df_pop_by_state_sort
```

```python
top_5_state_list = df_pop_by_state_sort['state'].tolist()
print('list of 5 states:' , top_5_state_list)
```

```python
state_str = ','.join(top_5_state_list)
print('list of states as one string:', state_str)
```
